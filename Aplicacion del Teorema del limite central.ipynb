# 2. Simularemos 1000 lanzamientos de dicho dado

```python
# Este código simula 1000 lanzamientos del dado
n = 1000

# En cada simulación, hay un ensayo más que en la simulación anterior
avg = []
for i in range(2,n):
    a = np.random.randint(1,7,i)
    avg.append(np.average(a))
```

```python
# Este codigo muestra 10 valores esperados de los lanzamientos
avg[1:10]
```

# 3. Función para trazar histograma y animación.

```python
# Este código ejecita la función que graficará el histograma, donde la corriente es la última figura.
def clt(current):
    plt.cla()
    if current == 1000: 
        a.event_source.stop()

    plt.hist(avg[0:current])

    plt.gca().set_title('Valor esperado al lanzar el dado')
    plt.gca().set_xlabel('Promedio de lanzada')
    plt.gca().set_ylabel('Frecuencia')

    plt.annotate('Lanzar dado = {}'.format(current), [3,27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, clt, interval=1)
```
